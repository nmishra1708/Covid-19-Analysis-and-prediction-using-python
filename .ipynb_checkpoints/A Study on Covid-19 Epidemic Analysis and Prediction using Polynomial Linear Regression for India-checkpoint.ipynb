```python
'''
About of Polynomial Regression;
Polynomial Regression is a form of linear regression in which the relationship between the independent variable x and dependent variable y is modeled as an nth degree polynomial. Polynomial regression fits a nonlinear relationship between the value of x and the corresponding conditional mean of y, denoted 
E(y |x) ……………………………………………………………………………………..(1)

Why Polynomial Regression:
•	There are some relationships that a researcher will hypothesize is curvilinear. Clearly, such type of cases will include a polynomial term.
•	Inspection of residuals. If we try to fit a linear model to curved data, a scatter plot of residuals (Y axis) on the predictor (X axis) will have patches of many positive residuals in the middle. Hence in such situation it is not appropriate.
•	An assumption in usual multiple linear regression analysis is that all the independent variables are independent. In polynomial regression model, this assumption is not satisfied.

Uses of Polynomial Regression:
These are basically used to define or describe non-linear phenomenon such as:
•	Growth rate of tissues.
•	Progression of disease epidemics
•	Distribution of carbon isotopes in lake sediments

The basic goal of regression analysis is to model the expected value of a dependent variable y in terms of the value of an independent variable x. In simple regression, we used following equation –
y = a + bx + e …………………………………………………………………….(2)

Here y is dependent variable, a is y intercept, b is the slope and e is the error rate.
y = a+ b1x + b2^2 + e…………………………………………………………….(3)

Here y is dependent variable on x, a is y intercept and e is the error rate.
In general, we can model it for nth value.
y = a + b1x + b2x^2 + ……+ bnx^n……………………………………………………(4) 

Since regression function is linear in terms of unknown variables, hence these models are linear from the point of estimation.
Since regression function is linear in terms of unknown variables, hence these models are linear from the point of estimation.

Advantages of using Polynomial Regression:
•	Broad range of function can be fit under it.
•	Polynomial basically fits wide range of curvature.
•	Polynomial provides the best approximation of the relationship between dependent and independent variable.

Disadvantages of using Polynomial Regression
•	These are too sensitive to the outliers.
•	The presence of one or two outliers in the data can seriously affect the results of a nonlinear analysis.
•	In addition there are unfortunately fewer model validation tools for the detection of outliers in nonlinear regression than there are for linear regression.

'''
```

### _Step(1)-> Pull Data from Source_

```python
#Step(1.1)-> Importing necessary Library
import pandas as pd

#Step(1.2)-> Pull data from data source
confirmedCases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deathsCases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recoveredCases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
updatedData = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-02-2020.csv')
```

```python
#Step(1.3)-> take a glance of Confirmed Cases
confirmedCases.head()

#Step(1.3)-> take a glance of Death Cases
deathsCases.head()

#Step(1.3)-> take a glance of Confirmed Cases
recoveredCases.head()

#Step(1.3)-> take a glance of Confirmed Cases
updatedData.head()
```

```python
#Step(1.4)->Extract all the labels from dataset
labels = confirmedCases.keys()

#Step(1.5)->Extract data for confirmed, death and recoveries
confirmed = confirmedCases.loc[:, labels[4]:labels[-1]]
deaths = deathsCases.loc[:, labels[4]:labels[-1]]
recoveries = recoveredCases.loc[:, labels[4]:labels[-1]]

#Step(1.6)->Extract the dates from dataset
dates = confirmed.keys()
```

```python
#Step(1.7)->Initialize the list of analytical attributes
worldCases = []
totalDeaths = [] 
mortalityRate = []
recoveryRate = [] 
totalRecovered = [] 
totalActive = [] 
indiaCases = [] 
```

```python
#Step(1.8)->Calculated values required for analysis

for date in dates:
    # Calculate total count of confirmed case, death cases and recovered cases
    confirmedTotal = confirmed[date].sum()
    deathTotal = deaths[date].sum()
    recoveredTotal = recoveries[date].sum()
    
    # confirmed, deaths, recovered, and active cases 
    worldCases.append(confirmedTotal)
    totalDeaths.append(deathTotal)
    totalRecovered.append(recoveredTotal)
    totalActive.append(confirmedTotal-deathTotal-recoveredTotal)
    
    # calculate rates of mortality and recovered cases
    mortalityRate.append(deathTotal/confirmedTotal)
    recoveryRate.append(recoveredTotal/confirmedTotal)
    
    # case studies for India
    indiaCases.append(confirmedCases[confirmedCases['Country/Region']=='India'][date].sum())
```

```python
#Step(1.9)-> Implement function for define dailyIncreased rate
#Daily increased report
def dailyIncrease(data):
    d = [] 
    for i in range(len(data)):
        if i == 0:
            d.append(data[0])
        else:
            d.append(data[i]-data[i-1])
    return d

world_daily_increase = dailyIncrease(worldCases)
india_daily_increase = dailyIncrease(indiaCases)
```

```python
#Step(1.10)-> Reshaping data for analysis purpose
import numpy as np
days_since_1_22 = np.array([i for i in range(len(dates))]).reshape(-1, 1)
world_cases = np.array(worldCases).reshape(-1, 1)
total_deaths = np.array(totalDeaths).reshape(-1, 1)
total_recovered = np.array(totalRecovered).reshape(-1, 1)

# Future Forecasting
days_in_future = 20
future_forcast = np.array([i for i in range(len(dates)+days_in_future)]).reshape(-1, 1)
adjusted_dates = future_forcast[:-20]
```

```python
#Step(1.11)->Convert integer into datetime for better visualization
import datetime
import time
start = '1/22/2020'
start_date = datetime.datetime.strptime(start, '%m/%d/%Y')
future_forcast_dates = []
for i in range(len(future_forcast)):
    future_forcast_dates.append((start_date + datetime.timedelta(days=i)).strftime('%m/%d/%Y'))
```

### _Step(2) -> Implement polynomial linear regression for prediction_

```python
#Step(2.1)->Separate the training and test set 80:20 ratio
from sklearn.model_selection import train_test_split
X_train_confirmed, X_test_confirmed, y_train_confirmed, y_test_confirmed = train_test_split(days_since_1_22, world_cases, test_size=0.25, shuffle=False)
```

```python
#Step(2.2)-> transform our data for polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=5)
poly_X_train_confirmed = poly.fit_transform(X_train_confirmed)
poly_X_test_confirmed = poly.fit_transform(X_test_confirmed)
poly_future_forcast = poly.fit_transform(future_forcast)
```

```python
#Step(2.3)-> Apply polynomial feature in linearRegression 
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression(normalize=True, fit_intercept=False)
linear_model.fit(poly_X_train_confirmed, y_train_confirmed)
test_linear_pred = linear_model.predict(poly_X_test_confirmed)
linear_pred = linear_model.predict(poly_future_forcast)
```

```python
#Step(2.4)-> Find the errors
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('MAE:', mean_absolute_error(test_linear_pred, y_test_confirmed))
print('MSE:',mean_squared_error(test_linear_pred, y_test_confirmed))
print('Coefficient:',linear_model.coef_)
```

### Step(3)->Visulalization the prediction

```python
#Step(3.1)-> Plot the comparison visualization between Test Data and Polynomial Regression Predictions

#Step(3.1.1)->Importing necessary library
import matplotlib.pyplot as plt
%matplotlib inline

#Step(3.1.2)->Add figure size
fig = plt.figure(figsize=(10,5))

#Step(3.1.3)-> add axes
ax1 = plt.subplot()

#Step(3.1.4)->Fit the x and y axis to the plot method
plt.plot(y_test_confirmed, linewidth=2.0, linestyle='--', color='b', alpha=0.5, marker='o')
plt.plot(test_linear_pred, linewidth=1.0, linestyle='--', color='r', alpha=1, marker='o')

#Step(3.1.5)-> Add the legend
plt.legend(['Test Data', 'Polynomial Regression Predictions'], loc='best')

#Step(3.1.6)->Add the grid
plt.grid(True)

#Step(3.1.7)-> Add the customized title and x and y label
plt.title('Commparison between Test Data Vs Polynomial Regression Predictions ')
plt.xlabel('Dates from 1-22-20')
plt.ylabel('Total no of Cases')

#Step(3.1.8)->Showing the plot
plt.show()
```

```python
#Step(3.2)->Plot Chart for future forecasting of no. of cases using Polynomial Regression Predictions
#Add the size
plt.figure(figsize=(10,5))
plt.plot(adjusted_dates, world_cases)
plt.plot(future_forcast, linear_pred, linestyle='dashed', color='red')
plt.title('Total no.of covid-19 Cases Over Time')
plt.xlabel('Days from 22-Jan-2020')
plt.ylabel('Total no. of Cases')
plt.legend(['Confirmed Cases', 'Polynomial Regression Predictions'])
plt.grid(True)
plt.show()
```
